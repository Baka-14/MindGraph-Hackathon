```python
import pandas as pd 
from matplotlib import pyplot as plt

df_combine=pd.read_csv("Combine.csv") 
df_combine.head()
```

```python
fest2=[]
name_fest2=[]
for i in range(1,16):
    fest2.append(df_combine['Event'].value_counts()[f"fest_2_event_{i}"])
    name_fest2.append(f"fest_1_event_{i}") 
   

fest1=[]
name_fest1=[]
for i in range(1,14):
    fest1.append(df_combine['Event'].value_counts()[f"fest_1_event_{i}"])
    name_fest1.append(f"fest_1_event_{i}") 
```

```python
df_fest1=pd.DataFrame({"Event Name": name_fest1,"Occurance": fest1})
plt.title('Fest 1')
plt.rcParams['figure.facecolor'] = 'white'
plt.pie(df_fest1["Occurance"], labels = df_fest1["Event Name"],autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2,radius=2.5)
plt.show()
plt.savefig('Fest1')

```

```python
df_fest2=pd.DataFrame({"Event Name": name_fest2,"Occurance": fest2})
plt.title('Fest 2')
plt.rcParams['figure.facecolor'] = 'white'
plt.pie(df_fest2["Occurance"], labels = df_fest2["Event Name"],autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2,radius=2.5)
plt.show()
plt.savefig('Fest2')
```

```python
club1=[]
names_club1=[]

club2=[]
names_club2=[]

club3=[]
names_club3=[]

df_combine.drop(df_combine[df_combine['Role']!="Participant"].index, inplace = True)
```

```python

for i in range(1,4):
    # print(f"club_1_event_{i}")
    club1.append(df_combine['Event'].value_counts()[f"club_1_event_{i}"])
    names_club1.append(f"club_1_event_{i}") 

for i in range(1,4):
    club2.append(df_combine['Event'].value_counts()[f"club_2_event_{i}"])
    names_club2.append(f"club_2_event_{i}") 

for i in range(1,4):
    club3.append(df_combine['Event'].value_counts()[f"club_3_event_{i}"])
    names_club3.append(f"club_3_event_{i}") 

del club1[3:]
del club2[3:]
del club3[3:]

del names_club1[3:]
del names_club2[3:]
del names_club3[3:]
```

```python
df_club1=pd.DataFrame({"Event Name": names_club1,"Occurance": club1})
plt.title('Club 1')
plt.rcParams['figure.facecolor'] = 'white'
plt.pie(df_club1["Occurance"], labels = df_club1["Event Name"],autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2,radius=2.65)
plt.show()
plt.savefig('Club 1')

df_club2=pd.DataFrame({"Event Name": names_club2,"Occurance": club2})
plt.title('Club 2')
plt.rcParams['figure.facecolor'] = 'white'
plt.pie(df_club2["Occurance"], labels = df_club2["Event Name"],autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2,radius=2.65)
plt.show()
plt.savefig('Club 2')

df_club3=pd.DataFrame({"Event Name": names_club3,"Occurance": club3})
plt.title('Club 3')
plt.rcParams['figure.facecolor'] = 'white'
plt.pie(df_club3["Occurance"], labels = df_club3["Event Name"],autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2,radius=2.65)
plt.show()
plt.savefig('Club 3')
```

```python
df_club_count = df_combine.groupby('Club_Name')['Roll No.'].nunique()

clubs = ["Club_1", "Club_2", "Club_3"]
club_values = [df_club_count[0],df_club_count[1],df_club_count[2]]
plt.bar(clubs, club_values, color ='black',width = 0.4)
plt.xlabel("Clubs")
plt.ylabel("No. of students Participated")
plt.show()
```

```python
df_roll_count_17X = df_combine["Roll No."].str.count("17XJ1").sum()
df_roll_count_18X = df_combine["Roll No."].str.count("18XJ1").sum()
students = ["17XJ", "18XJ"]
student_part_count = [df_roll_count_17X, df_roll_count_18X]
plt.bar(students, student_part_count, color ='black',width = 0.4)
plt.xlabel("Students")
plt.ylabel("No. of students Participated")
plt.show()
```
