```python
from difflib import SequenceMatcher as SM
```

```python
def is_similarity(first, second, ratio):
    return SM(None, first, second).ratio() > ratio
```

```python
for Index, row in df_organiser.iterrows():
    for Index2, row2 in df_meta.iterrows():
        if is_similarity(row["Name"], row2["Name"], 0.63):
            df_organiser.at[Index, "Name",] = row2["Name"]
df_organiser.head()
```

```python

club1=[]
names_club1=[]

club2=[]
names_club2=[]

club3=[]
names_club3=[]

df_club.drop(df_club[df_club['Role']!="Participant"].index, inplace = True)
df_club.shape

```
